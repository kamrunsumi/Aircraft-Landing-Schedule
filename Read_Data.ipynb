```python
import numpy as np
import os
import csv
```

```python



#Define function 
def read_file(file_name):  # filename is the name of the data file, i.e. airland1.txt
    data=open(os.getcwd()+'\\data_files\\'+file_name,'r') # Read the file from the subfolder "data_files"
    lines=data.readlines()  # read the file line by line
    num_planes=int(lines[0].split()[0]) # Get the first data from the first line which is number of planes
    freeze_time =int(lines[0].split()[1]) # Second data in the first line is the freeze time
    flight_details=np.empty([num_planes,6],dtype=float) # Define empty matrix for Appearance time, Ei, Ti, Li, gi, hi.
    sep_time=np.empty([num_planes,num_planes],dtype=int) # Define empty matrix for separation time no of planes x num f planes
    s=''  # Define empty string
    for line in lines[1:]:  # read from line second since first line is already read in the above paragraph
        s=s+line # read all the lines
    s=s.split()  # split each line
    flag=0
    count=0
    for items in [s[x:x+6+num_planes] for x in range(0,len(s),num_planes+6)]:
        flight_details[count]=[float(x) for x in items[:6]] # items[:6] includes appearance time, Ei, Ti, LI, gi, hi
        sep_time[count]=[int(x) for x in items[6:]] # separation time starts from the column which has index 6. The previous line read upto index no 5
        count=count+1 ## We need to collect data for 10 planes (num_planes)
    print(num_planes)
    print(freeze_time)
    print(flight_details)
    print(sep_time)
    data.close() # it is a good practice to close the the file which is opened earlier. 
    return num_planes,freeze_time,flight_details,sep_time
```

```python
fil={1:1,2:8, 3:9, 4:13}
#For airland1.txt with 10 flights
num=4
#num_flights,freeze_time,flights,clearance=read_file('airland'+str(fil[num])+'.txt')
num_flights,freeze_time,flights,clearance=read_file('airland'+str(12)+'.txt')
#print(fil[3])
```

```python
#print (flights[0,5])
#print (num_flights())

tcol=2+6+num_flights
trow=num_flights

final_data=np.empty([trow,tcol],dtype=float) 

for i in range(trow):
    final_data[i][0]=float(i+1)
    final_data[i][1]=float(freeze_time)
for i in range(trow):
    for j in range(6):
        final_data[i][j+2]=flights[i][j]
    
for i in range(trow):
    for j in range(num_flights):
        final_data[i][j+8]=clearance[i][j]

sdata=np.array_split(final_data,tcol,axis=0)
#rint (final_data.shape)
#rint (final_data[0])
print(sdata)
#rint(type(final_data))
```
